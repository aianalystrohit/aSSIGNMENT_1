```python
# Labeling data fields to Text value for easy interpretation of Visualization
data_eda = df.copy()

data_eda["sex"]                 = df["sex"] .map({1: "Male", 0: "Female"})
data_eda["diabetes"]            = df["diabetes"].map({1: "Yes", 0: "No"})
data_eda["anaemia"]             = df["anaemia"].map({1: "Yes", 0: "No"})
data_eda["high_blood_pressure"] = df["high_blood_pressure"].map({1: "Yes", 0: "No"})
data_eda["smoking"]             = df["smoking"].map({1: "Yes" , 0: "No"})
data_eda["DEATH_EVENT"]         = df["DEATH_EVENT"].map({1: "Yes", 0: "No"})
```

```python
data_eda
```

```python
filtered_data = data_eda[data_eda['DEATH_EVENT'] != 'No']  # Replace 'Y' with your actual value

plt.figure(figsize=(10, 6))
sns.violinplot(x='age', y='DEATH_EVENT', data=filtered_data)

plt.ylabel('Death Event (Yo)')

plt.xlabel('Age')
plt.title("Violin Plot: Age Distribution for 'Yes' Death Events")
plt.grid(True)
plt.show()
```

```python
#Death rates are higher during 55 to 65 and minimal after 85+
```
